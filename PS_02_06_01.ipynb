# 二项分布PMF
《概率统计》 | 鸢尾花书：数学不难

这段代码的核心任务是从数学上定义一个**二项分布**的概率质量函数（PMF），并通过可视化将其呈现出来。整个过程严格遵循了概率论中对离散型随机变量分布的描述方法。

1.  **参数设定与随机变量的定义**
    首先，代码设定了二项分布的两个关键参数：试验次数 $n = 30$ 和单次试验的成功概率 $p = 0.5$。对于一个服从二项分布的随机变量 $X$，其定义为 $X \sim \text{Binomial}(n=30, p=0.5)$。这意味着 $X$ 表示在 30 次独立的伯努利试验中成功的总次数。$X$ 是一个离散型随机变量，其所有可能的取值构成了一个状态空间，即从 0 次成功到 30 次成功的所有整数，数学上表示为 $x \in \{0, 1, 2, \dots, 30\}$。代码中通过 `np.arange(0, n+1)` 生成了这一组离散的样本点。

2.  **概率质量函数（PMF）的计算**
    针对上述每一个可能的取值 $x$，代码利用 `scipy.stats.binom.pmf` 计算了该特定值发生的精确概率。在数学上，二项分布的概率质量函数 $p_X(x)$ 定义为：
    $$
    p_X(x) = P(X = x) = \binom{n}{x} p^x (1-p)^{n-x}, \quad \text{for } x = 0, 1, 2, \dots, n
    $$
    代入本例中的具体参数 $n=30$ 和 $p=0.5$，该公式变为：
    $$
    P(X = x) = \binom{30}{x} (0.5)^x (0.5)^{30-x} = \binom{30}{x} (0.5)^{30}
    $$
    代码对每一个 $x$ 计算了对应的概率值，从而得到一组离散的概率值 $\{P(X=0), P(X=1), \dots, P(X=30)\}$。从公式中可以看出，当 $p=0.5$ 时，分布是对称的，因为 $(0.5)^x (0.5)^{30-x} = (0.5)^{30}$ 是一个常数，概率的大小完全由组合数 $\binom{30}{x}$ 决定。

3.  **概率分布的图形化表示**
    为了直观地展示该分布，代码采用了**茎叶图（stem plot）** 进行绘制。这种图形非常适合表示离散型随机变量的分布：
    - **横轴** 对应随机变量的取值 $x$，范围为 $[-1, 31]$，确保所有样本点都在视野内。
    - **纵轴** 对应每个 $x$ 的概率 $p_X(x)$。代码将纵轴上限设置为 `max(pmf)*1.1`，这相当于在最大概率密度处预留了约 10% 的空白，以避免图形顶到边界。
    - 图中每一个 “茎” 的顶点坐标对应于点 $(x, P(X=x))$。这些点的集合 $(x_i, p_X(x_i))$ 完整地刻画了该二项分布的概率结构。由于 $p=0.5$，图形会呈现出以 $x=15$（即 $np$）为中心的对称形态。

## 初始化

```python
import numpy as np                                      # 导入 numpy 库用于数值计算
import matplotlib.pyplot as plt                         # 导入 matplotlib.pyplot 用于绘图
from scipy.stats import binom                           # 导入二项分布
```

## 二项分布

```python
n = 30                                                  # 试验次数
p_success = 0.5                                         # 成功概率

x = np.arange(0, n+1)                                   # 二项分布取值范围 0,...,n
pmf = binom.pmf(x, n, p_success)                        # 使用 scipy.stats.binom 计算 PMF
```

## 绘图

```python
plt.figure(figsize=(8, 4))                              
markerline, stemlines, baseline = plt.stem(x, pmf, 
                                           linefmt='#0066FF', markerfmt='#0066FF', basefmt='#0066FF')     
plt.setp(markerline, marker='o', markersize=6)                       
plt.setp(baseline, visible=False)                      
    
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"PMF, $p_{X}(x)$", fontsize=12)
plt.grid(True, color='0.8', alpha=0.5)
plt.xlim(-1, n+1)
plt.ylim(0, max(pmf)*1.1)                               # 根据实际PMF值调整y轴范围
plt.xticks(np.arange(0, n+1, 2))                         # 每2个单位显示一个刻度
    
plt.grid(axis='y', linestyle='--', linewidth=0.5)      
plt.title(f'Binomial Distribution PMF (n={n}, p={p_success})', fontsize=14)
plt.tight_layout()                                       # 自动调整布局
```
