# 模拟抛一颗色子10次
《概率统计》 | 鸢尾花书：数学不难

这段代码从数学角度模拟了**离散型随机实验**——连续掷一枚公平六面骰子的过程，并记录每次的实验结果。
在概率论中，这一过程可以形式化描述为：

1. **随机试验定义**
   设随机试验为掷一枚标准六面骰子，样本空间为
   $$ S = {1, 2, 3, 4, 5, 6} $$
   其中每个样本点表示骰子朝上的点数。

2. **随机变量定义**
   定义随机变量$X$表示一次掷骰子的结果，则
   $$ X: S \to {1,2,3,4,5,6}, \quad P(X = k) = \frac{1}{6}, ; k = 1,2,3,4,5,6 $$
   这说明$X$服从**离散均匀分布** (Discrete Uniform Distribution)。

3. **重复独立试验**
   代码中的循环

   ```python
   for i in range(10):
       roll = random.randint(1, 6)
   ```

   数学上对应进行10次独立重复试验，每次生成一个独立同分布的随机变量$X_i$，且
   $$ X_1, X_2, \dots, X_{10} \overset{i.i.d.}{\sim} U{1,2,3,4,5,6} $$
   每个$X_i$的期望为
   $$ E[X_i] = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = 3.5 $$
   方差为
   $$ \mathrm{Var}(X_i) = \frac{(6^2 - 1)}{12} = \frac{35}{12} \approx 2.9167 $$

4. **结果表示**
   代码将每次的结果存入一个列表 `results`，从数学上相当于构成一个样本集合：
   $$ {x_1, x_2, \dots, x_{10}} $$
   其中每个$x_i$是一次独立掷骰得到的样本值。

5. **输出解释**
   最终输出的`results`列表代表该随机试验在10次独立重复中的**具体观测样本序列**。
   这是一种从理论分布 $U{1,2,3,4,5,6}$ 中得到的**经验样本（empirical sample）**。
   如果多次重复实验并计算平均值
   $$ \bar{X} = \frac{1}{10}\sum_{i=1}^{10} X_i $$
   根据**大数定律**，当试验次数趋向无穷大时，有
   $$ \bar{X} \to E[X] = 3.5 $$
   即实验平均值会逐渐逼近期望值。

简而言之，这段代码用计算机的伪随机数生成过程，数值化地再现了掷骰子的数学模型，通过重复独立采样展示了离散均匀分布在有限样本下的一个具体实现。


## 初始化

````python
import random  
# 导入随机模块
````

## 抛一枚色子10次

````python
results = []  # 用来保存每次的结果
for i in range(10):
    roll = random.randint(1, 6)  
    # 随机生成1到6之间的整数，模拟掷一次色子
    results.append(roll)  
    # 将结果加入列表
````

## 输出所有掷骰结果

````python
print("Dice roll results:")
print(results)
````
