# 判断对立事件
《概率统计》 | 鸢尾花书：数学不难

1. **定义样本空间与事件**
   代码首先给出样本空间和两个事件：
   $$\Omega={1,2,3,4,5,6},\qquad A={1,3,5},\qquad B={2,4,6}.$$
   这里 $A$ 表示“出现奇数”，$B$ 表示“出现偶数”。

2. **检验互斥（交集为空）**
   代码通过判断 `A & B == set()` 来检测
   $$A\cap B=\varnothing.$$
   在本例中，确实有
   $${1,3,5}\cap{2,4,6}=\varnothing,$$
   所以 $A$ 与 $B$ 互斥（不能同时发生）。

3. **检验完备（并集覆盖样本空间）**
   代码通过判断 `(A | B) == Omega` 来检测
   $$A\cup B=\Omega.$$
   代入得到
   $${1,3,5}\cup{2,4,6}={1,2,3,4,5,6}=\Omega,$$
   因此两事件的并集覆盖了整个样本空间（没有遗漏的基本结果）。

4. **用补集定义验证对立关系**
   代码还用 `A_complement = Omega - A` 并判断 `B == A_complement`，这对应数学上的补集关系：
   $$A^{c}=\Omega\setminus A={2,4,6}.$$
   若 $B=A^{c}$，则 $B$ 是 $A$ 的补集，二者互为对立事件。

5. **对立事件的等价条件**
   在集合与概率论中，事件 $A$ 与 $B$ 互为对立（complementary）当且仅当同时满足两条：
   $$
   \begin{cases}
   A\cap B=\varnothing,[4pt]
   A\cup B=\Omega.
   \end{cases}
   $$
   代码分别用交集为空与并集等于 $\Omega$ 的两个判断实现了这一定义。

6. **概率角度的简单验证（均匀样本空间）**
   若假设色子每一面等可能，则每一结果概率为 $\dfrac{1}{6}$。记 $|\cdot|$ 为集合基数，有 $|A|=3,\ |B|=3,\ |\Omega|=6$，于是
   $$P(A)=\frac{|A|}{|\Omega|}=\frac{3}{6}=\frac{1}{2},\qquad P(B)=\frac{3}{6}=\frac{1}{2}.$$
   因而
   $$P(A)+P(B)=1,$$
   与 $B=A^{c}$ 的概率含义一致。

## 定义样本空间与事件

```python
Omega = {1, 2, 3, 4, 5, 6}   
# 样本空间
A = {1, 3, 5}                 
# 事件A：奇数
B = {2, 4, 6}                
# 事件B：偶数
```

## 判断对立

```python
# 判断集合交集是否为空，即判断互斥
A & B == set()
```

```python
# 判断集合的并集是否覆盖样本空间
(A | B) == Omega
```

```python
# 也可以判断B是否等于A的补集
A_complement = Omega - A 
# A的补集
B == A_complement
```
