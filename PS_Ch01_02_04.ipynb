### 1. 样本空间与事件定义

* 该代码隐含的随机试验是一个**离散型随机试验**（例如掷一枚标准骰子）。

* 样本空间定义为：
  $$
  \Omega = {1,2,3,4,5,6}
  $$

* 在此基础上定义两个事件：
  $$
  A = {1,2}, \quad B = {5,6}
  $$

* 从概率语义上看：

  * $A$ 表示“点数小于 3”
  * $B$ 表示“点数大于 4”

---

### 2. 事件的集合运算（交集）

* 代码中的：

  ```
  A & B
  ```

  对应集合论中的**交集运算**：
  $$
  A \cap B
  $$

* 实际计算结果为：
  $$
  A \cap B = \varnothing
  $$

* 这意味着：

  * 不存在任何样本点同时属于 $A$ 和 $B$
  * 即：
    $$
    \forall \omega \in \Omega,\quad \omega \notin A \cap B
    $$

---

### 3. 互斥事件的数学定义

* 代码中的：

  ```
  A.isdisjoint(B)
  ```

  本质是在判断：
  $$
  A \cap B = \varnothing
  $$

* 在概率论中，这正是**互斥事件（mutually exclusive events）**的定义：
  $$
  A \cap B = \varnothing ;\Longleftrightarrow; A \text{ 与 } B \text{ 互斥}
  $$

---

### 4. 概率层面的含义

* 若假设为均匀分布（标准骰子），则：
  $$
  P(A) = \frac{2}{6}, \quad P(B) = \frac{2}{6}
  $$

* 由于 $A$ 与 $B$ 互斥，有：
  $$
  P(A \cap B) = 0
  $$

* 因此满足加法公式（互斥特例）：
  $$
  P(A \cup B) = P(A) + P(B)
  $$

---

### 5. 数学本质总结

* 该代码的核心在于验证以下等价关系：
  $$
  A.isdisjoint(B) ;\Longleftrightarrow; A \cap B = \varnothing
  $$

* 从结构上看，它完成了三件事：

  * 用集合表示随机事件
  * 用集合运算刻画事件关系
  * 用“空交集”刻画概率论中的互斥性
