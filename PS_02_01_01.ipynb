# 离散均匀分布PMF
《概率统计》 | 鸢尾花书：数学不难

这段代码的核心任务，是对一个**离散均匀分布**（discrete uniform distribution）进行数学建模，并绘制它的**概率质量函数**（Probability Mass Function, PMF）。我们考虑的随机变量为掷一颗标准六面色子的结果。

---

## 1. 随机变量与样本空间

代码首先隐含地定义了一个离散随机变量
$$
X : \Omega \rightarrow {1,2,3,4,5,6},
$$
它表示掷一颗公平六面色子后的点数。样本空间为
$$
\mathcal{X} = {1,2,3,4,5,6}.
$$

由于色子是公平的，因此每个样本点的概率完全相同，即
$$
\mathbb{P}(X = k) = \frac{1}{6}, \quad k = 1,2,3,4,5,6.
$$

代码中的

```python
x = np.arange(1, 7)
pmf = np.full_like(x, fill_value=1/6, dtype=float)
```

正是数学上将这组离散点 ${1,2,3,4,5,6}$ 与各自的概率 $\frac{1}{6}$ 一一对应。

---

## 2. PMF 的数学结构

概率质量函数 $p_X(x)$ 由代码构造为：

$$
p_X(x) =
\begin{cases}
\frac16, & x \in {1,2,3,4,5,6}, \
0, & \text{其他 } x.
\end{cases}
$$

这是离散均匀分布的典型形式，其特点包括：

* 所有概率相等：$p_X(x_1)=p_X(x_2)$ 对任意 $x_1, x_2 \in \mathcal{X}$ 都成立；
* 满足规范化条件：
  $$
  \sum_{x\in\mathcal{X}} p_X(x) = 6\times \frac{1}{6} = 1.
  $$

代码将其可视化为**扑棱图（stem plot）**，即在每个整数点 $x=k$ 处绘制一条从 $0$ 到 $1/6$ 的竖线，这正是离散 PMF 的图像。

---

## 3. 图像意义：将离散概率用图形表示

绘图指令

```python
plt.stem(x, pmf, ...)
```

从数学上就是在可视化函数 $p_X(x)$ 的值。每一根竖线代表“在点 $x=k$ 上的概率大小”，即：

* 竖线的高度 = $p_X(k)=\frac16$
* 横坐标刻度正是样本空间中的每个离散点。

此外，图中添加了一条水平参考虚线：

$$
y = \frac{1}{6},
$$

用于强调“每个点的概率完全相同”，这是离散均匀分布的典型特征。

---

## 4. 可视化区间与格式设置的数学含义

代码中设置

```python
plt.xlim(0, 7)
plt.ylim(0, 0.2)
```

从数学意义上，这只是将 PMF 函数在更大的区域中展示，使读者可以直观看出该函数仅在六个整数点处非零，其余点的概率均为 0，即：

$$
p_X(x) = 0, \quad x \notin {1,2,3,4,5,6}.
$$

这强调了 PMF 的**离散性**（仅在离散点定义）和**稀疏性**（除了有限个点，其余处皆为零）。

---

## 5. 整体数学结构总结

这段代码从数学角度的核心是：

1. 设定离散随机变量 $X$，其取值在 $\mathcal{X}={1,\dots,6}$。
2. 构造离散均匀分布的 PMF：
   $$
   p_X(x)=\frac{1}{6}.
   $$
3. 使用可视化方法展示函数 $p_X : \mathcal{X}\rightarrow \mathbb{R}$ 的形状，使概率分布更直观。
4. 图像揭示两个数学事实：

   * **均匀性**：所有 $p_X(k)$ 相等。
   * **离散性**：只有在整数点才有概率，非整数点的函数值全部为 0。


## 初始化

````python
import numpy as np
import matplotlib.pyplot as plt
````

## 生成数据

````python
x = np.arange(1, 7)                                     
pmf = np.full_like(x, fill_value=1/6, dtype=float)     
````

## 可视化

````python
plt.figure(figsize=(6,3))                              
markerline, stemlines, baseline = plt.stem(x, pmf, 
                                           linefmt='#0066FF', 
                                           markerfmt='#0066FF', 
                                           basefmt='#0066FF')     
plt.setp(markerline, marker='o')                       
plt.setp(baseline, visible=False)                      
plt.axhline(y = 1/6, c = 'r', ls = '--')
plt.xlabel(r"$x$")
plt.ylabel(r"PMF, $p_{X}(x)$")
plt.grid(True, color = '0.8')
plt.xlim(0, 7)
plt.xticks(np.arange(0,8))
plt.ylim(0, 0.2)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', linewidth=0.5)                                          
````
