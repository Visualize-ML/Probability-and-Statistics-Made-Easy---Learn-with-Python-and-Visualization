# 模拟抛两枚硬币，连续抛10次
《概率统计》 | 鸢尾花书：数学不难

从数学角度来看，这段代码模拟了**两枚硬币独立抛掷的随机实验**，并重复进行了 $10$ 次。每次实验包含两个独立的伯努利随机变量。整个过程可以理解为生成一组来自二维离散随机分布的独立样本。

1. 设第 $i$ 次实验中，两枚硬币的结果分别用随机变量 $X_i$ 和 $Y_i$ 表示：
   $$
   X_i =
   \begin{cases}
   1, & \text{若第一枚硬币为正面 (Head)} \
   0, & \text{若第一枚硬币为反面 (Tail)}
   \end{cases},
   \quad
   Y_i =
   \begin{cases}
   1, & \text{若第二枚硬币为正面 (Head)} \
   0, & \text{若第二枚硬币为反面 (Tail)}
   \end{cases}
   $$
   每个变量都独立且服从参数为 $p=0.5$ 的伯努利分布，即
   $$P(X_i=1)=P(X_i=0)=0.5, \quad P(Y_i=1)=P(Y_i=0)=0.5.$$

2. 由于两枚硬币相互独立，因此它们的联合分布为
   $$
   P(X_i=x, Y_i=y) = P(X_i=x) \cdot P(Y_i=y) = 0.25,
   $$
   其中 $x, y \in {0, 1}$。
   这意味着每次抛掷的四种可能结果（HH, HT, TH, TT）出现的概率都是 $0.25$。

3. 代码中的循环执行 $10$ 次，即产生了 $10$ 个独立样本：
   $$
   {(X_1, Y_1), (X_2, Y_2), \ldots, (X_{10}, Y_{10})}.
   $$
   每个样本代表一次双硬币抛掷的结果，整个列表 `results` 就是这些随机向量的观测集合。

4. 如果我们关心“正面总数”的随机变量，可以定义
   $$S_i = X_i + Y_i,$$
   则 $S_i$ 的取值为 ${0, 1, 2}$，并且服从二项分布
   $$S_i \sim \text{Binomial}(n=2, p=0.5).$$
   其概率质量函数为
   $$
   P(S_i = k) = \binom{2}{k} (0.5)^2, \quad k = 0, 1, 2.
   $$

5. 因此，从数学角度看，这段程序实际上在**生成 10 个来自二项分布 $B(2, 0.5)$ 的样本对**，并输出每次实验的两个独立结果 $(X_i, Y_i)$。每次运行程序，相当于重新从相同的联合分布中抽取一组独立样本。


## 初始化

```python
import random  # 导入随机模块
```

## 抛两枚硬币，连续抛10次

```python
results = []  # 用来保存每次抛掷的结果
for i in range(10):
    toss1 = random.choice(['Head', 'Tail'])  # 第一枚硬币
    toss2 = random.choice(['Head', 'Tail'])  # 第二枚硬币
    results.append((toss1, toss2))  # 存成一个元组，加入列表
```

## 输出所有抛掷结果

```python
print("Double coin toss results:")
for idx, (t1, t2) in enumerate(results):
    print(f"Trial {idx+1}: {t1}, {t2}")
```
