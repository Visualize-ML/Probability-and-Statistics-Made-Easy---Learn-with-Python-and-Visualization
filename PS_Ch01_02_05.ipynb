# 差事件
《概率统计》 | 鸢尾花书：数学不难

在这段代码中我们考察集合运算中的**差集（差事件）**，也就是“在事件 $A$ 发生但不在事件 $B$ 中”的那些结果。

$A = {2,4,6}$ 表示“点数为偶数”，$B = {4,5,6}$ 表示“点数大于 3”。差集的数学定义为：

$$
A\setminus B={x\mid x\in A\ \text{且}\ x\notin B}
$$

根据定义逐项判断：$2\in A$ 且 $2\notin B$，所以 $2$ 属于 $A\setminus B$；$4\in A$ 但 $4\in B$，因此 $4$ 不属于差集；$6\in A$ 且 $6\in B$，也不属于差集。由此得到具体结果：

$$
A\setminus B={2}
$$

差集也可以用交集和补集来表示，记 $B^c$ 为全集（此处可取骰子可能的所有面 ${1,\dots,6}$）中不属于 $B$ 的元素，则有等价表达：

$$
A\setminus B = A\cap B^c
$$

在概率层面（若以均匀骰子为例），差事件的概率满足减法规则：若 $P$ 是概率测度，则

$$
P(A\setminus B)=P(A)-P(A\cap B)
$$

在 Python 中，`A - B` 与 `A.difference(B)` 两种写法等价，都是计算上述差集 $A\setminus B$ 并返回新的集合 ${2}$。


## 定义事件A和事件B

```python
A = {2, 4, 6}       
# 事件A：点数为偶数
B = {4, 5, 6}      
# 事件B：点数大于3
```

## 计算差事件

```python
A - B  
```

```python
# 也可以写成 
A.difference(B)
```
