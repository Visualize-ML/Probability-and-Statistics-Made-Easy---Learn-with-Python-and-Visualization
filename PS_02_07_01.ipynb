# 泊松分布PMF
《概率统计》 | 鸢尾花书：数学不难

该段代码的核心任务是**从一个参数为 $\lambda = 4$ 的泊松分布中，计算其概率质量函数，并将该函数以离散茎干图的形式进行可视化**。整个过程可以分解为以下三个数学步骤：

### 1. 定义随机变量的支撑集
首先，代码确定了随机变量 $X$ 的取值范围。由于泊松分布的定义域为所有非负整数，但概率质量主要集中在均值附近，代码选取了从 $0$ 到 $20$ 的整数作为绘图范围：
$$
\text{support} = \{ x \in \mathbb{Z} \mid 0 \leq x \leq 20 \}
$$
在代码中，使用 `np.arange(0, 21)` 生成了这个离散集合 $x = [0, 1, 2, \dots, 20]$。

### 2. 计算泊松分布的概率质量函数
给定泊松分布的参数 $\lambda = 4$，对于支撑集中的每一个 $x$，计算其对应的概率质量。根据泊松分布的定义，概率质量函数 $p_X(x)$ 的数学表达式为：
$$
p_X(x) = P(X = x) = \frac{e^{-\lambda} \lambda^x}{x!}
$$
代入 $\lambda = 4$，代码实际计算的公式为：
$$
\text{pmf}[x] = \frac{e^{-4} \cdot 4^x}{x!}, \quad \text{for } x = 0, 1, 2, \dots, 20
$$
该计算由 `scipy.stats.poisson.pmf(x, lambda_i)` 完成，其输出是一个长度为 21 的向量，包含了所有 $x$ 对应的概率值。

### 3. 概率分布的图形化呈现
为了直观展示该离散概率分布，代码采用了**茎干图**进行绘制。图中的每一个茎干（垂直线段）代表了在特定 $x$ 处的概率质量大小。图中点的纵坐标严格等于第二步计算出的概率值，即：
$$
\text{点的纵坐标} = p_X(x) = \frac{e^{-4} \cdot 4^x}{x!}
$$
绘图时，横轴范围被限定在 $[-0.5, 20.5]$ 以确保所有数据点完整显示，纵轴范围则根据 $\lambda = 4$ 时的理论最大值（在 $x=3$ 和 $x=4$ 附近）设定为 $[0, 0.25]$，因为对于 $\lambda = 4$，其最大概率值满足：
$$
\max_{x} p_X(x) \approx p_X(4) = \frac{e^{-4} \cdot 4^4}{24} \approx 0.195
$$
位于 0.25 的限幅之内。

## 初始化

```python
import numpy as np                                      # 导入 numpy 库用于数值计算
import matplotlib.pyplot as plt                         # 导入 matplotlib.pyplot 用于绘图
from scipy.stats import poisson                         # 导入泊松分布
```

```python
lambda_i = 4
    
x = np.arange(0, 21)
pmf = poisson.pmf(x, lambda_i)
```
