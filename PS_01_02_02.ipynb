在这段代码中，我们研究的是两个事件 $A$ 和 $B$ 之间的“并集”关系。事件 $A$ 被定义为掷出点数小于 $4$ 的结果，因此 $A = {1, 2, 3}$；事件 $B$ 被定义为掷出偶数点数的结果，因此 $B = {2, 4, 6}$。

从数学角度来看，两个事件的**并集**表示“至少发生一个事件”的情形。即如果掷出的结果属于 $A$ 或属于 $B$（或同时属于两者），就认为并事件发生。我们用符号表示为

$$
A \cup B = {x \mid x \in A \text{ 或 } x \in B}
$$

代码 `union = A | B` 正是用集合运算符“|”计算这个并集。Python 中的集合并运算与数学定义完全一致，会返回一个新的集合，包含 $A$ 和 $B$ 中的所有不同元素。根据定义：

$$
A = {1, 2, 3}, \quad B = {2, 4, 6}
$$

因此它们的并集为

$$
A \cup B = {1, 2, 3, 4, 6}
$$

这意味着，当掷骰子的结果是 $1, 2, 3, 4,$ 或 $6$ 时，就属于事件 $A \cup B$ 的范围。换句话说，这个并事件表示“点数小于 4 或为偶数”。

代码的最后一行 `A.union(B)` 是另一种写法，功能与 `$A | B$` 完全相同。函数 `.union()` 是集合对象的一个内置方法，用于返回两个集合的并集，因此数学上两者都表示相同的运算：

$$
A \cup B = A.\text{union}(B)
$$


## 定义事件A和事件B

```python
A = {1, 2, 3}         
# 事件A：点数小于4
B = {2, 4, 6}         
# 事件B：点数为偶数
```

## 计算和事件 (并集)

```python
union = A | B  
```

```python
# 也可以写成 
A.union(B)
```
