# 包含
《概率统计》 | 鸢尾花书：数学不难

这段代码从数学角度描述了**事件之间的包含关系**（subset relation），它以掷一颗标准六面骰子的样本空间为背景，形式化地演示了事件$A$与事件$B$之间的集合关系。

---

1. **样本空间的构建**
   在一次掷骰子的随机试验中，样本空间为
   $$ S = {1, 2, 3, 4, 5, 6} $$
   每个元素代表可能出现的点数。所有事件都是$S$的子集。

---

2. **定义事件$A$与事件$B$**
   代码中定义了：
   $$
   A = {6}, \quad B = {2, 4, 6}
   $$
   其中，事件$A$表示“掷出6点”，事件$B$表示“掷出偶数”。
   在集合论框架下，事件的逻辑关系都可以通过集合运算（如交、并、补、包含）来描述。

---

3. **事件的包含关系（Subevent）**
   判断$A$是否为$B$的子事件，即：
   $$
   A \subseteq B
   $$
   这在概率论中表示：若事件$A$发生，则事件$B$必然同时发生。换句话说，$A$的发生蕴含了$B$的发生。
   在此例中，掷出6点（事件$A$）必然意味着掷出了一个偶数（事件$B$），因此确实有
   $$
   {6} \subseteq {2, 4, 6}
   $$

---

4. **代码判断的数学意义**
   表达式`A.issubset(B)`或`A <= B`在Python集合运算中，分别对应数学上的：
   $$
   A \subseteq B
   $$
   该语句会返回布尔值`True`或`False`，即判断事件$A$是否完全包含于事件$B$之中。

---

5. **概率角度的解释**
   若样本空间$S$中每个结果等可能（即均匀分布），则事件的概率为集合大小与样本空间大小的比值：
   $$
   P(A) = \frac{|A|}{|S|}, \quad P(B) = \frac{|B|}{|S|}
   $$
   在此例中：
   $$
   P(A) = \frac{1}{6}, \quad P(B) = \frac{3}{6} = \frac{1}{2}
   $$
   因为$A \subseteq B$，所以根据概率的单调性：
   $$
   P(A) \le P(B)
   $$

---

综上，这段代码在数学上展示了**事件包含关系的逻辑与概率含义**。它不仅验证了集合上的$A \subseteq B$，也对应了概率论中“事件$A$发生蕴含事件$B$必然发生”的关系。


## 定义事件A和事件B

```python
A = {6}            
# 事件A：掷出6
B = {2, 4, 6}      
# 事件B：掷出偶数
```

## 判断包含关系

```python
A.issubset(B)
```

```python
# 也可以写成
A <= B
```
