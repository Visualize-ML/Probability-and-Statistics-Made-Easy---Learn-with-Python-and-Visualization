# 积事件
《概率统计》 | 鸢尾花书：数学不难

从数学角度来看，这段代码主要是在定义两个事件并计算它们的**积事件（交集）**，即同时满足两个事件条件的样本点。具体分析如下：

首先，定义事件集合：

* 设样本空间为掷一颗六面色子的所有可能结果：
  $$S = {1, 2, 3, 4, 5, 6}$$
* 事件 $A = {1, 2, 3}$ 表示“点数小于 4”的结果。
  因此在集合形式上，$A \subset S$。
  其数学含义为：
  $$A = {x \in S \mid x < 4}$$
* 事件 $B = {2, 4, 6}$ 表示“点数为偶数”的结果。
  同样地，$B \subset S$，可写作：
  $$B = {x \in S \mid x \text{ 是偶数}}$$

接着，代码计算两个事件的积事件（intersection）：

* 在集合论中，两个事件的积事件 $A \cap B$ 定义为：
  $$A \cap B = {x \mid x \in A \text{ 且 } x \in B}$$
  即“同时属于事件 $A$ 和事件 $B$ 的样本点集合”。
* 将上面的定义代入：
  $$A = {1, 2, 3}, \quad B = {2, 4, 6}$$
  因此：
  $$A \cap B = {2}$$

这表示只有当色子点数为 $2$ 时，两个条件（点数小于 4 且为偶数）同时成立。

从概率角度进一步解释，若色子是均匀的（每个点数出现的概率相等），则每个样本点的概率为
$$P({x}) = \frac{1}{6}$$
因此积事件的概率为：
$$P(A \cap B) = \frac{|A \cap B|}{|S|} = \frac{1}{6}$$

综上，这段代码的核心是演示**集合交集操作**在概率事件中的含义：
事件 $A$ 与事件 $B$ 的交集代表“同时发生”的情形，计算结果 ${2}$ 表示该条件仅在点数为 2 时满足。


## 定义事件A、事件B

```python
A = {1, 2, 3}       
# 事件A：点数小于4
B = {2, 4, 6}       
# 事件B：点数为偶数
```

## 计算积事件 (交集)

```python
A & B  
```

```python
A.intersection(B)
```
